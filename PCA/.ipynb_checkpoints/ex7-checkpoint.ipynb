```python
# 人脸降维
data = io.loadmat("/home/sjl/桌面/machine learning/mlclass-ex7-005/mlclass-ex7/ex7faces.mat")
X = data['X']
```

```python
# 显示图像
def display(X):
    m, n = X.shape
    exampleWidth = round(math.sqrt(n)) # 样本的宽度
    exampleHeight = int(n/exampleWidth) # 样本的高度
    
    displayRows = round(math.sqrt(m)) # 显示的行数
    displayCols = int(m/displayRows) # 显示的列数
    pad = 1 # 两两图像之间的间距
    
    displayArray = np.zeros((pad+displayRows*(pad+exampleHeight), pad+displayCols*(pad+exampleWidth))) # 显示图像的矩阵
    for i in range(displayRows):
        for j in range(displayCols):
            displayArray[i*(exampleHeight+pad)+pad:(i+1)*(exampleHeight+pad), j*(exampleWidth+pad)+pad:(j+1)*(exampleWidth+pad)]=X[i*displayCols+j].reshape(exampleHeight, exampleWidth)
    plt.imshow(displayArray.T, cmap='gray') 
    plt.show()
display(X[:100])
```

```python
# 运行pca算法, 根据前100个主成分进行pca算法
p = PCA(X, 100)
w, v, Z = p.pca()
recoverX = p.recoverData()
```

```python
# 展示图像, 展示主成分特征与恢复后的数据
display(v.T[:36])
display(recoverX[:36])
```
